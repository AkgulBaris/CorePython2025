# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
giris_sayisi = int(input("Kac kisi icin veri girisi yapacaksiniz "))

with open("vki_kayitlari.txt", "a", encoding="utf-8") as dosya:
    for i in range(giris_sayisi):
        print(f"\n{i+1}. kişi için bilgileri girin:")

        ad_soyad = input("Ad Soyad: ")
        yas = input("Yas: ")
        boy = float(input("Boy (metre): "))
        kilo = float(input("Kilo (kg): "))

        vki = kilo / (boy ** 2)

        if vki < 18.5:
            sinif = "Zayif"
        elif vki <= 24.9:
            sinif = "Normal kilolu"
        elif vki <= 29.9:
            sinif = "Fazla kilolu"
        elif vki <= 39.9:
            sinif = "Obez"
        else:
            sinif = "Ileri derecede obez"

        print(f"{ad_soyad} adli kisinin VKİ degeri: {vki:.2f} → {sinif}")

        dosya.write(f"{ad_soyad}, Yas: {yas}, Boy: {boy} m, Kilo: {kilo} kg, VKI: {vki:.2f}, Sinif: {sinif}\n")
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import time

menu = {
    "elma": 0.50,
    "uzum": 0.75,
    "armut": 0.90,
    "cilek": 0.95,
    "muz": 1,
    "yabanmersini": 1.50,
    "ananas": 2
}

print("Elma Fiyati:", menu["elma"])
print("Ananas Fiyati:", menu["ananas"])

def display_menu():
    print("---YetBar Menu---")
    for item in list(menu.keys()):
        print(f"{item.upper():<10}: {menu[item]} TL")

def tek_urun_fiyat_hesaplama(urun, adet):
    assert adet >= 0, "Urun adedi 0 ya da 0'dan kucuk olamaz"
    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun] * adet
    
def satin_alma():
    display_menu()
    print("Lutfen satin almak istediginiz urunu giriniz. Cikis yapmak için 0 tusuna basiniz")

    satin_alinan_urunler = []
    satin_alma_adedi = []
    son_fiyat = 0

    while True:
        eklenecek_urun = input("Hangi urunden eklemek istiyorsunuz? ")
        if eklenecek_urun == "0":
            print("Fis kaydediliyor.")
            break
        ekleme_adedi = int(input("Kac adet eklemek istiyorsunuz? "))
        fiyat = tek_urun_fiyat_hesaplama(eklenecek_urun,ekleme_adedi)
        if fiyat is None:
            print("Yanlis girdiniz. Urun ekleme basarisiz olmustur.")
        else:
            print(f"Satin alma ucreti: {fiyat} TL.\nSatin almayi onaylamak icin 'y', iptal icin 'x' giriniz.")
            check = input("Satin almayi onayliyor musunuz? ")
            if check == "y":
                print("Satin aliminiz onaylandi")
                print("-----")
                son_fiyat += fiyat
                satin_alinan_urunler.append(eklenecek_urun)
                satin_alma_adedi.append(ekleme_adedi)
                print(f"Ara Toplam: {son_fiyat} TL ")
            elif check == "x":
                print("Satin alma islemi iptal edildi")
                print("-----")
    return son_fiyat, satin_alinan_urunler, satin_alma_adedi

def main():
    print("YetBar'a Hos Geldiniz!\nMenuyu Goruntulemek icin '1' e, smoothie yapmak icin '2' ye basabilirsiniz. Cikmak icin '0' a basiniz. ")
    program = None
    while True:
        program = int(input("Tercihiniz: "))
        if program < 0 or program > 2:
            print("Yanlis bir tercih yaptiniz. Lutfen yeniden deneyiniz.")
        elif program == 0:
            print("Cikis yapiliyor.\nTesekkurler!")
            break
        elif program == 1:
            display_menu()
        elif program == 2:
            satin_alma_ucreti, satin_alinan_urunler, satin_alma_adedi = satin_alma()
            with open("kasa.txt", "a", encoding="utf-8") as file:
                file.write(str(time.ctime(time.time())))
                file.write("\n")
                file.write(','.join(map(str, satin_alinan_urunler)))
                file.write("\n")
                file.write(','.join(map(str, satin_alma_adedi)))
                file.write("\n")
                file.write(f"Toplam: {satin_alma_ucreti} TL")
                file.write("\n")
                file.write("-----")
                file.write("\n")
            cikis_kontrol = input("Baska bir islem yapmak ister misiniz? (evet/hayir) ")
            if cikis_kontrol.lower() =="evet":
                pass
            else:
                print("\n\nTesekkurler")
                break

if __name__ == " __main__":
    main()




                

```
